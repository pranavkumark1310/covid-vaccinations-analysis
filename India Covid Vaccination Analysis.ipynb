### Dropping column 'Unamed: 0' and analysing the whole dataset using lineplot.

```python
india_vaccine_data.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
sns.lineplot(data=india_vaccine_data, palette="tab10", linewidth=2.5)
```

### Vaccination trends in India between 10-04-2021 and 13-04-2021

```python
plt.figure(figsize=(16,10))
new_data = india_vaccine_data[(india_vaccine_data['date'] >= '10-04-2021') & (india_vaccine_data['date'] <= '17-04-2021')]
plt.plot(new_data['date'], new_data['total_vaccinations']/1000)
```

```python
sns.pairplot(data=new_data)
```

### Line graph for total vaccinations vs People Fully vaccinated

```python
plt.figure(figsize=(16,10))
plt.plot(new_data['total_vaccinations']/1000, new_data['people_fully_vaccinated']/1000)
plt.title('Total vaccinations vs People Fully vaccinated')
plt.xlabel('Total Vaccinations')
plt.ylabel('People Fully Vaccinated')
plt.show()
```

```python
plt.figure(figsize=(16,10))
plt.plot(new_data['people_vaccinated_per_hundred']/1000, new_data['people_fully_vaccinated_per_hundred']/1000)
plt.title('Vaccinated per hundred vs Fully vaccinated per hundred')
plt.xlabel('Vaccinations per hundred')
plt.ylabel('Fully Vaccinated per hundred')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(16, 16))
sns.scatterplot(x=india_vaccine_data.total_vaccinations/1000, y=india_vaccine_data.people_fully_vaccinated/1000,
                palette="ch:r=-.2,d=.3_r",
                sizes=(1, 10), linewidth=0,
                data=india_vaccine_data,ax=ax)
```

```python
plt.figure(figsize=(20,10))
sns.barplot(new_data.date, new_data.daily_vaccinations/1000, palette='CMRmap_r')
plt.title("Vaccination in India")
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(new_data.date, new_data.daily_change_in_vaccinations/1000)
```
